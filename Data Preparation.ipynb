```python
train_df, val_df = train_test_split(train_df, stratify=train_df[target])
```

# Signup date

```python
col = 'signup_date'
train_df[col] = pd.to_datetime(train_df[col])
val_df[col] = pd.to_datetime(val_df[col])
test_df[col] = pd.to_datetime(test_df[col])
```

# created_at

```python
col = 'created_at'
train_df[col] = pd.to_datetime(train_df[col])
val_df[col] = pd.to_datetime(val_df[col])
test_df[col] = pd.to_datetime(test_df[col])

train_df[col+'_month'] = train_df[col].dt.month
val_df[col+'_month'] = val_df[col].dt.month
test_df[col+'_month'] = test_df[col].dt.month
```

```python
cnts = train_df[col+'_month'].value_counts(dropna=False, normalize=True)
cnts.cumsum()
```

```python
ct = pd.crosstab(train_df[col+'_month'], train_df[target])
ct.divide(ct.sum(axis=1), axis='rows')*100
```

```python
train_df['signup_date_presence'] = train_df['signup_date'].isna().astype(int)
val_df['signup_date_presence'] = val_df['signup_date'].isna().astype(int)
test_df['signup_date_presence'] = test_df['signup_date'].isna().astype(int)
```

```python
train_df['diff'] = (train_df['created_at'] - train_df['signup_date']).dt.days
val_df['diff'] = (val_df['created_at'] - val_df['signup_date']).dt.days
test_df['diff'] = (test_df['created_at'] - test_df['signup_date']).dt.days
```

```python
train_df['diff'] = train_df['diff'].fillna(9999)
val_df['diff'] = val_df['diff'].fillna(9999)
test_df['diff'] = test_df['diff'].fillna(9999)
```

```python
cols = ['campaign_var_1_trim', 'campaign_var_2_trim', 'user_act_trim']
for col in cols:
    dummies = pd.get_dummies(train_df[col])
    dummies.columns = col+'_'+dummies.columns.astype(str)
    train_df = pd.concat([train_df, dummies], axis=1)
    
    dummies = pd.get_dummies(val_df[col])
    dummies.columns = col+'_'+dummies.columns.astype(str)
    val_df = pd.concat([val_df, dummies], axis=1)

    dummies = pd.get_dummies(test_df[col])
    dummies.columns = col+'_'+dummies.columns.astype(str)
    test_df = pd.concat([test_df, dummies], axis=1)
```

```python
cols = ['campaign_var_1_trim', 'campaign_var_2_trim', 'user_act_trim', 'products_purchased', 'created_at_month']
for col in cols:
    freq = train_df[col].value_counts(dropna=False, normalize=True)
    train_df[col+'_freq'] = train_df[col].map(freq)
    val_df[col+'_freq'] = val_df[col].map(freq).fillna(0)
    test_df[col+'_freq'] = test_df[col].map(freq).fillna(0)
```

```python
for col in cols:
    tme = train_df.groupby(col)[target].mean()
    train_df[col+'_tme'] = train_df[col].map(tme)
    val_df[col+'_tme'] = val_df[col].map(tme).fillna(0)
    test_df[col+'_tme'] = test_df[col].map(tme).fillna(0)
```

```python
drop_cols = [
    'created_at', 'signup_date', 
#     'campaign_var_1', 'campaign_var_2',

#     'user_activity_var_1', 'user_activity_var_2', 'user_activity_var_3',
#     'user_activity_var_4', 'user_activity_var_5', 'user_activity_var_6',
#     'user_activity_var_7', 'user_activity_var_8', 'user_activity_var_9',
#     'user_activity_var_10', 'user_activity_var_11','user_activity_var_12',

    'user_act',
    'user_act_trim', 'campaign_var_1_trim', 'campaign_var_2_trim'
]
```

```python
train_df = train_df.drop(columns=drop_cols, errors='ignore')
val_df = val_df.drop(columns=drop_cols, errors='ignore')
test_df = test_df.drop(columns=drop_cols, errors='ignore')
```
